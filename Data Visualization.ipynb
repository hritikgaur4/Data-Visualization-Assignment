Q.10 Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
np.random.seed(42)
size = 100  # Number of samples

# Generate correlated data
x1 = np.random.normal(0, 1, size)
x2 = 2 * x1 + np.random.normal(0, 0.1, size)
x3 = -x1 + np.random.normal(0, 0.1, size)
x4 = 0.5 * x1 + np.random.normal(0, 0.1, size)

# Create a DataFrame
data = pd.DataFrame({'Feature 1': x1, 'Feature 2': x2, 'Feature 3': x3, 'Feature 4': x4})

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)

# Adding title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Adding title and labels
ax.set_title('3D Surface Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(25)

# Generate the dataset
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5, c='blue')

# Adding title and labels
plt.title('Bubble Chart of Country Population and GDP')
plt.xlabel('GDP')
plt.ylabel('Population')

# Annotate each point with the country name
for i in range(df.shape[0]):
    plt.text(df['GDP'][i], df['Population'][i], df['Country'][i], fontsize=9, ha='right')

# Display the plot
plt.show()
```

Q.16 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set up Bokeh to display plots in the notebook
output_notebook()

# Generate the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

Q.17 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Set up Bokeh to display plots in the notebook
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 50, size=100)
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], size=100)

# Create a DataFrame
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a new plot
p = figure(title="Scatter Plot with Different Sizes and Colors", x_axis_label='X', y_axis_label='Y')

# Add a scatter renderer with different sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, alpha=0.6)

# Show the plot
show(p)
```

Q.18 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.  
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Set up Bokeh to display plots in the notebook
output_notebook()

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

Q.20 Create a Bokeh histogram to visualize the distribution of the given data.  
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Set up Bokeh to display plots in the notebook
output_notebook()

# Generate the data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

# Create a new plot
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad renderer with the histogram data
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color='navy', line_color='white', alpha=0.7)

# Show the plot
show(p)
```

Q.21 Create a Bokeh heatmap using the provided dataset.  
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Set up Bokeh to display plots in the notebook
output_notebook()

# Generate the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for Bokeh
data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools='hover', tooltips=[("Value", "@value")])

# Add a rect renderer
p.rect(x='x', y='y', width=1/10, height=1/10, source=source,
       fill_color=linear_cmap('value', Viridis256, 0, 1), line_color=None)

# Show the plot
show(p)
```
